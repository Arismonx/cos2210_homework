```python
class PhoneBattery:
    def __init__(self):
        self.__percent = 100
    
    def use_battery(self,amount):
        if amount < self.__percent:
            self.__percent -= amount
        else:
            self.__percent = 0
    
    def charge_battery(self,amount):
        if self.__percent + amount > 100:
            self.__percent = 100
        else:
            self.__percent += amount 
    
    def get_Battery(self):
        return self.__percent
    
phone = PhoneBattery()
phone.use_battery(20)
print(phone.get_Battery())

phone.use_battery(30)
print(phone.get_Battery())

phone.use_battery(60)
print(phone.get_Battery())

phone.charge_battery(60)
print(phone.get_Battery())

phone.charge_battery(50)
print(phone.get_Battery())

phone.use_battery(80)
print(phone.get_Battery())
```
