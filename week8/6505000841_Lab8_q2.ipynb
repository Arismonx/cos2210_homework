```python
class BankAccount:
    
    def __init__(self,account_number,balance = 0):
        self.account_number = account_number
        self._balance = balance

    def deposit(self,amount):
        if amount > 0:
            self._balance += amount
            print(f"ฝาก {amount} บัญชี {self.account_number} ยอด {self._balance}")
        else:
            print(f"จํานวนเงินต้องมากกว่า 0")
            
    def withdraw(self,amount):
        if amount > 0 & amount < self._balance:
            self._balance -= amount
            print(f"ถอน {amount} บัญชี {self.account_number} ยอด {self._balance}")
        else:
            print(f"ยอดเงินไม่พอหรือจํานวนเงินไม่ถูกต้อง")
    
    def get_balance(self):
        return self._balance
    

class SavingsAccount(BankAccount):
    interest_rate = 0.015
    
    def __init__(self, account_number, balance = 0):
        super().__init__(account_number, balance)
    
    def add_interest(self):
        interest = self._balance * SavingsAccount.interest_rate
        self._balance += interest
        print(f"เพิ่มดอกเบี้ย {interest:.2f} บัญชี {self.account_number}")
```

```python
mysaving = SavingsAccount("123-456-789")

mysaving.deposit(10000)
print(mysaving.get_balance())

mysaving.add_interest()
print(mysaving.get_balance())

mysaving.withdraw(2000)
print(mysaving.get_balance())
```
