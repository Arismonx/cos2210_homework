````python
distance = input("Enter travel distance (km.):")
type_car = input("Enter the type of car:")
pattern_drive = input("Enter driving pattern:")

print(f"Enter travel distance (km.): {distance}")
print(f"Enter the type of car: {type_car}")
print(f"Enter driving pattern: {pattern_drive}")

distance = float(distance)

if type_car == "1":
    if pattern_drive == "1":
        avgUseOli = 12.0
    elif pattern_drive == "2":
        avgUseOli = 10.0
elif type_car == "2":
    if pattern_drive == "1":
        avgUseOli = 10.0
    elif pattern_drive == "2":
        avgUseOli = 8.0
elif type_car == "3":
    if pattern_drive == "1":
        avgUseOli = 9.0
    elif pattern_drive == "2":
        avgUseOli = 7.0
        
fuel = distance/avgUseOli

print(f"Fuel required: {fuel:.2f}")
````
