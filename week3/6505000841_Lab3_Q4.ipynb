checksum เป็นการคํานวณค่าเพื่อตรวจสอบความถูกต้องของข้อมูล โดยมาก
checksum จะเป็นการเติมเลข 1 หลักต่อท้ายค่าที่ต้องการ หากผู้ใช้ป้อนข้อมูลผิดพลาด
ค่า checksum ที่คํานวณได้จะเปลี่ยนไป ทําให้ตรวจสอบในเบื้องต้นได้ว่า
การป้อนข้อมูลถูกต้องหรือไม่ จงเขียนโปรแกรมเพื่อคํานวณหาค่า checksum
เมื่อผู้ใช้ป้อนเลขประจําตัวประชาชน (13 หลัก) เข้ามาให้ระบบ เช่น 3795357249057
จะมี checksum (หลักแรก ขวาสุด) เป็น 7 ดังตัวอย่างข้างล่างนี้

ตัวอย่าง
```
--- ID Checksum Calculator ---
Enter 13-digit ID number: 3795357249057
Calculated checksum digit 7
```
การคํานวณ checksum ในเลขประจําตัวประชาชนเป็นดังนี้
- (a) นําเลขแต่ละหลักมาคูณกับเลขประจําหลัก (หลักที่ 13 ถึงหลักที่ 2
เรียงจากซ้ายไปขวา)
- (b) หาผลบวกของจํานวนในข้อ (a) (12 จํานวน)
- (c) นําผลลัพธ์ในข้อ (b) หารด้วย 11 เก็บค่าเศษเหลือเอาไว้
- (d) หาผลต่างของ 11 กับเศษเหลือที่ได้ในข้อ (c)
- (e) นําหลักหน่วยของผลลัพธ์ในข้อ (d) มาเป็นเลขหลักสุดท้าย (หลักที่ 1)
ของเลขประจําตัวประชาชน

````python
id_number = input("Enter 13-digit ID number:")

base = 13
a = 0.0
b = 0.0
c = 0.0
d = 0.0

print(f"Enter 13-digit ID number: {id_number}")

for i in range(12):
    a = float(id_number[i]) * base
    base -= 1
    b += a
    c = b % 11
    d = 11 - c

print(f"Calculated checksum digit: {int(d)}")
        
````
