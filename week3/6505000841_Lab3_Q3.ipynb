จงเขียนโปรแกรมเครื่องมือตรวจคําตอบข้อสอบปรนัย โดยโปรแกรมมีคุณสมบัติดังต่อไปนี้
- รับสตริงเฉลยและสตริงคําตอบนักเรียนต่อเนื่อง
- เมื่อผู้ใช้พิมพ์ 'exit' (ไม่สนตัวพิมพ์เล็ก-ใหญ่) ในช่องสตริงเฉลย
โปรแกรมจะจบการทํางานแล้วแสดงข้อความว่า 'Exiting program.
Goodbye!'
- เปรียบเทียบคําตอบแต่ละข้อแบบไม่สนตัวพิมพ์เล็ก - ใหญ่ (เช่น 'A' เท่ากับ 'a')
แล้วนับจํานวนข้อที่ถูก แสดงข้อความว่า "Number of correct
answers: [จํานวนข้อที่ถูก]"
- ถ้าจํานวนข้อของเฉลยไม่ตรงกับของคําตอบ ให้แสดง Incomplete answer

ตัวอย่าง
```
--- Multiple Choice Answer Checker ---
Enter the string of correct answers: ABABC
Enter the string of student's answers: ABBBC
Number of correct answers: 4
Enter the correct answers (type 'exit' to quit): XYZ
Enter the student's answers: xy
Incomplete answer
Enter the correct answers (type 'exit' to quit): exit
Exiting program. Goodbye!
```

````python
def CheckLenAndCheckAns(correct_ans,student_ans):
    len_correct_ans = len(correct_ans)
    len_student_ans = len(student_ans)
    count = 0
    if  len_correct_ans == len_student_ans:
        for i in range(len_correct_ans):
            if correct_ans[i].lower() == student_ans[i].lower():
                count += 1
        print(f"Number of correct answers: {count}")
    else:
        print("Incomplete answer")
````

````python
# Main Process
print("--- Multiple Choice Answer Checker ---")
correct_ans = input("Enter the string of correct answers:")
student_ans = input("Enter the string of student's answers:")

print(f"Enter the string of correct answers: {correct_ans}")
print(f"Enter the string of student's answers: {student_ans}")

CheckLenAndCheckAns(correct_ans,student_ans)


while 1:
    correct_ans = input("Enter the correct answers (type 'exit' to quit):")
    print(f"Enter the correct answers (type 'exit' to quit): {correct_ans}")
    
    if correct_ans.lower() == "exit":
        break

    student_ans = input("Enter the student's answers:")
    print(f"Enter the student's answers: {student_ans}")
    
    CheckLenAndCheckAns(correct_ans,student_ans)
     
print("Exiting program. Goodbye!")
````
