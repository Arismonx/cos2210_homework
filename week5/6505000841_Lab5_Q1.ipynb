```python
import math

def distance1(x1, y1, x2, y2):
    return math.sqrt(((x1-x2)**2) + ((y1-y2)**2))

def distance2(p1, p2):
    return math.sqrt(((p1[0]-p2[0])**2) + ((p1[1]-p2[1])**2))

def distance3(c1, c2):
    d = distance2(c1,c2)
    if d < c1[2]+c2[2]:
        # print("overlap")
        return d, True
    return d, False

def perimeter(points): # points[[0,0],[0,2],[2,2],[2,0]]
    total = 0
    n = len(points)
    for i in range(n):
        p1 = points[i]
        p2 = points[(i + 1) % n]
        d = distance2(p1,p2)
        total += d
    return total

print(distance1(0, 0, 3, 4))
print(distance2([0,0], [3,4]))
a,b = distance3([0,0,1], [5,0,2]); print(a, b)
print(perimeter([[0,0], [0,2], [2,2], [2,0]]))
```
