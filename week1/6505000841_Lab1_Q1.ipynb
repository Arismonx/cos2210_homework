## Q1
จงเขียนโปรแกรมที่รับราคาสินค้าและค่าขนส่งจากผู้ใช้ แล้วคำนวณยอดรวมที่ลูกค้าต้องจ่าย แล้วแสดง
ผลลัพธ์ดังตัวอย่างข้างล่างนี้

### ตัวอย่างการทำงาน:

Price : 235.50 <br>
Shipping :  30 <br>
Total: 265.5 Baht (30.0 Baht shipping fee)

### ตัวอย่างการทำงาน:

Price : 1000 <br>
Shipping :  45.5 <br>
Total: 1045.5 Baht (45.5 Baht shipping fee)


```python
price , shipping = input("Enter price and shipping:").split(",")
print("Price:",price)
print("Shipping:",shipping)
total = float(price) + float(shipping)
print(f"Total: {total} Baht ({shipping} Baht shipping fee)")
```
