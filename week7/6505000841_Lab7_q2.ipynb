```python
def convert_to_float(v):
    return [float(i) for i in v]

def read_file_vector(filename):
    v1 = []
    v2 = []
    with open(filename,"r") as f:
        v1 = [i for i in f.readline().split()]
        v2 = [i for i in f.readline().split()]
        
        v1 = convert_to_float(v1)
        v2 = convert_to_float(v2)
        
    return v1,v2

def is_equal(v1,v2):
    return len(v1)==len(v2)

def dot(v1,v2):    
    result = 0
    for i in range(len(v1)):
        result += v1[i] * v2[i]
    return result


files = ["vector1.txt", "vector2.txt", "vector3.txt", "vector4.txt"]

for fn in files:
    v1, v2 = read_file_vector(fn)
    
    print(f"Choose your vector file: {fn}")
    print(f"v1 = {v1}")
    print(f"v2 = {v2}")
    
    if is_equal(v1, v2):
        print(f"v1 * v2 : {dot(v1, v2)}")
    else:
        print("Incompatible size")
    
    print("========================")
```
