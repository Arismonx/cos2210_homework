```python
import math 

def quad1(a,b,c):
    if a == 0:
        raise Exception("Invalid problem")
    
    dis = b**2 - (4*a*c)
    if dis < 0:
        raise ValueError("Invalid problem")
    return (-b + math.sqrt(dis)) / (2*a)

def quad2(a,b,c):
    if a == 0:
        raise Exception("Invalid problem")
    
    dis = b**2 - (4*a*c)
    if dis < 0:
        raise ValueError("Invalid problem")
    return (-b - math.sqrt(dis)) / (2*a)


files = ["problems1.txt", "problems2.txt"]

for fn in files:
    print(f"Choose your problem file: {fn}")
    try:
        with open(fn,"r") as f:
            for line in f:
                try:
                    data = [float(i) for i in line.split()]
                    if not data: continue
                    
                    a,b,c = data
                    q1 = quad1(a,b,c)
                    q2 = quad2(a,b,c)
                    
                    print(f"{q1} {q2}")
                except Exception as e:
                    print(e)
                    
    except FileExistsError:
        print("File not found!")
        
    print("===========================")
```
