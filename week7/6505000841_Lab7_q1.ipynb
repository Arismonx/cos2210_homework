```python

def moveRobot(fn,x,y):
    with open(fn,"r") as f:
        
        nX,nY = x,y

        for line in f :
            if not "\n":
                continue
            if "L" == line[0]:
                nX -= 1
            elif "R" == line[0]:
                nX += 1
            elif "U" == line[0]:
                nY += 1
            elif "D" == line[0]:
                nY -= 1
            else:
                raise Exception("Invalid command")
    return nX,nY

files = ["move1.txt", "move2.txt"]

for fn in files:
    try:
        x , y = input("Enter x , y:").split()
        x , y = int(x),int(y)
        
        print(f"Choose your movefile: move1.txt")
        print(f"Initial position : {x},{y} ")
        
        nX,nY = moveRobot(fn,x,y)
        
        print(f"Robot stops at: {nX},{nY} ")
        print("===================")
    except FileExistsError:
        print("Cannot open file!")
    except Exception as e:
        print(e)
```
