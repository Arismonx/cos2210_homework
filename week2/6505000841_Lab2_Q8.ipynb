## Q8
จงเขียนโปรแกรมที่วนรับจำนวนเงินที่ต้องการเบิกจากบัญชี โดยจะยอมให้เบิกถ้ายังมีเงินในบัญชีพอ และวนรับจำนวนเงินที่ต้องการเบิกไปจนเงินในบัญชีหมดพอดี (กำหนดให้เริ่มต้นมีเงินในบัญชี 50,000 บาท)


### ตัวอย่างการทำงาน: 

withdraw : **50000** <br>
Balance is 0

### ตัวอย่างการทำงาน: 

withdraw : **45000**  <br>
withdraw : **5000**  <br>
Balance is 0

### ตัวอย่างการทำงาน: 

withdraw : **60000**  <br>
Insufficient fund. <br>
withdraw : **40000**  <br>
withdraw : **20000**  <br>
Insufficient fund.<br>
withdraw : **10000**  <br>
Balance is 0

```python
balance = 50000

while balance != 0:
    withdraw = input("withdraw:")
    withdraw = int(withdraw)
    
    print(f"withdraw : {withdraw}")
    
    if withdraw <= balance:
        balance = balance - withdraw
    else:
        print("Insufficient fund.")
            
print(f"Balance is {balance}")
```
