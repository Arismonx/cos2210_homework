### ตัวอย่างการทำงาน:

Number of students : **-1** <br>
Incorrect Input. Number of students : **7** <br>
Student 1 : **10** <br>
Student 2 : **6.5** <br>
Student 3 : **0** <br>
Student 4 : **9** <br>
Student 5 : **8.25** <br>
Student 6 : **9.50** <br>
Student 7 : **7.25** <br>
Total score : 50.5 <br>

```python
number = input("Number of students:")

while int(number) <= 0:
    number = input("Incorrect Input.Number of students:")

number = int(number)
total = 0.0
for i in range(number):
    cal = input(f"Studen {i+1} :")
    print(f"Studen {i+1} : {cal}")
    total += float(cal)
print(f"Total score: {total}")
```
