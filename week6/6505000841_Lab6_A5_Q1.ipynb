```python
def calculate_stats(*number):
    """
    Calculate sum avgrage max min keep in dictionary.
    
    Args: 
    *number(int or float): input multi value.
    
    Return:
    dictionary: Return dictionary.
    """
    sum = 0
    avg = 0
    max = 0
    min = 0
    
    dictionary = {
        'sum':sum,
        'average': avg,
        'max':max,
        'min':min
    }
    
    if not number:
        # print("test")
        return dictionary
    
    else:
    
        for i in number:
            sum += i
        
        avg = sum/len(number)
        
        for i in number:
            if i > max:
                max = i
        
        min = number[0]
        for i in number:
            if i < min:
                min = i
         
    
        dictionary = {
            'sum':sum,
            'average': avg,
            'max':max,
            'min':min
        }
    
        return dictionary
```

```python
print(calculate_stats(10, 20, 30, 40, 50))
print(calculate_stats(5, 5))
print(calculate_stats())

print(calculate_stats.__doc__)
```
