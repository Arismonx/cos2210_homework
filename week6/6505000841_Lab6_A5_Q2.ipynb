```python
def final_price(price,tax_rate=0.07,**discounts):
    """
    Calculate Net and check discounts.
    
    Args: 
    price: price product.
    default parameter tax_rate=0.07: vat7%.
    **discounts: dictionary parameter.
    
    Return:
    net: Return Net.
    """
    
    total_discount = 0.0
    sum = 0.0
    for name,dis in discounts.items():
        if name[:8] == "special_":
            dis *= 2
        if name[:8] == "expired_":
            dis = 0
        else:
            total_discount += dis
    
    sum = price - total_discount
    
    if sum < 0:
        sum = 0
    
    net = sum*(1+tax_rate)
    
    return net
```

```python
print(f"Case 1: {final_price(1000, discount_nov=100, expired_dec=500)}")
print(f"Case 2: {final_price(2000, special_vip=200)}")
print(f"Case 3: {final_price(3000, promo=100, special_year=200, expired_old=1000)}")
print(f"Case 4: {final_price(500, special_clearance=300)}")
print(f"Case 5: {final_price(2000, tax_rate=0, member=500)}")

print(final_price.__doc__)
```
