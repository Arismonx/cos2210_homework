```python
def power_recursive(base, exponent):
    """
    Find Power by use recursive.
    
    Args: 
    base: base number.
    exponent: number power.
    
    Return:
    base: Returns the number of a recursive function.
    """
    
    if exponent == 0:
        return 1
    else:
        return base * power_recursive(base,exponent-1)
```

```python
print(f"2^3 = {power_recursive(2, 3)}")
print(f"5^0 = {power_recursive(5, 0)}")
print(f"5^2 = {power_recursive(5, 2)}")

print(power_recursive.__doc__)
```
