```python
dictionary = []


while(True):
    print("--- Student Profile Management ---")
    print("1. Add Student")
    print("2. View Student")
    print("3. Update Student")
    print("4. Delete Student")
    print("5. View All Students")
    print("6. Exit")
    print("----------------------------------")
    opt = input("please choose an option:")
    print("please choose an option:",opt)
    
    if opt == '1':
        check = False
        if dictionary == []:
            student_id = input("Enter Studen ID:")
            name = input("Enter Full Name:")
            print("Enter Studen ID: ",student_id)
            print("Enter Full Name: ",name)
            dictionary.append([student_id,name])
            print("Student added successfully.")
        else:
            student_id = input("Enter Studen ID:")
            print("Enter Studen ID: ",student_id)
            for i in range(len(dictionary)):
                if student_id == dictionary[i][0]:
                    check = True
                    break
            if check:
                print("Error: Student ID already exists.")
            else:
                name = input("Enter Full Name:")
                print("Enter Full Name: ",name)
                dictionary.append([student_id,name])
                print("Student added successfully.")
        
    if opt == '2':
        check = True
        student_id = input("Enter Studen ID:")
        print("Enter Studen ID: ",student_id)
        for i in range(len(dictionary)):
            if student_id == dictionary[i][0]:
                print("Result:",dictionary[i][1])
                check = False
                break
        if check:
            print("Result: Student ID not found.")
        
    if opt == '3':
        check = True
        student_id = input("Enter Studen ID to update:")
        print("Enter Studen ID to update: ",student_id)
        for i in range(len(dictionary)):
            if student_id == dictionary[i][0]:
                name = input("Enter Full Name:")
                print("Enter Full Name: ",name)
                dictionary[i][1] = name
                print("Student data update successfully.")
                check = False
                break
        if check:
            print("Error: Student ID not found.")
    if opt == '4':
        check = True
        student_id = input("Enter Studen ID to delete:")
        print("Enter Studen ID to delete: ",student_id)
        for i in range(len(dictionary)):
            if student_id == dictionary[i][0]:
                del dictionary[i]
                print("Student data delete successfully.")
                check = False
                break
        if check:
            print("Error: Student ID not found.")
    if opt == '5':
        print("--- All Student ---")
        for i in range(len(dictionary)):
            print(f"ID: {dictionary[i][0]}, Name: {dictionary[i][1]}")
        print("-------------------")
    if opt == '6':
        print("Exiting program. Goodbye!")
        break
```
