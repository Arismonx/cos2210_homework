```python
points = [(5, 2), (3, 8), (4, 4), (3, 9), (25, 5), (10, 1), (2, 4), (9, 3)]

a = [ i[::-1] for i in points ]
b = [i for i in points if i[0] > i[1]]
c = [i for i in points if i[0] < i[1]]
d = [i for i in points if i[0] == i[1]]
e = [i for i in points if i[0]**2 == i[1]]
f = [i for i in points if i[1]**2 == i[0]]

print("a:",a)
print("b:",b)
print("c:",c)
print("d:",d)
print("e:",e)
print("f:",f)
```
