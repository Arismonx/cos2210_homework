```python
python = ['SUPANAT', 'TEERAPON', 'KAWINWAT', 'KIDSAKORN', 'PHON', 'SURACHAD']
java = ['PAWARIS', 'SUPANAT', 'PONGSAKORN', 'RATTANACHAT', 'PHON', 'SURACHAD']
cpp = ['SUPANAT', 'TEERAPON', 'KAWINWAT', 'RATTANACHAT', 'SURACHAD', 'MATCHAKAN']

result = []
for name in python:
    if name in java and name in cpp:
        result.append(name)
print(f"1. Enrolled in all 3 courses ({len(result)} students)")
print(f"   Name: {result}\n")

result = []
for name in python:
    if name not in java and name not in cpp:
        result.append(name)
print(f"2. Enrolled in only Python ({len(result)} students)")
print(f"   Name: {result}\n")

result = []
for name in java:
    if name not in python and name not in cpp:
        result.append(name)
print(f"3. Enrolled in only Java ({len(result)} students)")
print(f"   Name: {result}\n")

result = []
for name in cpp:
    if name not in python and name not in java:
        result.append(name)
print(f"4. Enrolled in only C++ ({len(result)} students)")
print(f"   Name: {result}\n")

result = []
for name in python:
    if name in java and name not in cpp:
        result.append(name)
print(f"5. Enrolled in Python and Java only ({len(result)} students)")
print(f"   Name: {result}\n")

result = []
for name in python:
    if name not in java and name in cpp:
        result.append(name)
print(f"6. Enrolled in Python and C++ only ({len(result)} students)")
print(f"   Name: {result}\n")

result = []
for name in java:
    if name not in python and name in cpp:
        result.append(name)
print(f"7. Enrolled in Java and C++ only ({len(result)} students)")
print(f"   Name: {result}\n")


result1 = []
result2 = []
result3 = []

for name in python:
    if name in java:
        result1.append(name)

for name in java:
    if name in cpp:
        result2.append(name)

for name in cpp:
    if name in python:
        result3.append(name)

all_result = result1 + result2 + result3

result = []
for name in all_result:
    if name not in result:
        result.append(name)

print(f"8. Enrolled in at least 2 courses ({len(result)} students)")
print(f"   Name: {result}\n")
```
