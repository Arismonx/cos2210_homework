```python
deck = []
commands = []

deck = input("Please enter the card in the deck (separated by spaces):").split()
commands = input("Please enter the commands (e.g., CSCS):").split()

half = len(deck) // 2
for i in range(len(commands)):
    if commands[i].lower() == 'c':
        deck = deck[half:] + deck[:half]
        
    if commands[i].lower() == 's':
        half1 = deck[:half]
        half2 = deck[half:]
        new_deck = []
        
        for i in range(half):
            new_deck.append(half1[i])
            new_deck.append(half2[i])
            
        deck = new_deck    
           
    if commands[i].lower() == 'x':
        break

print("The final state of deck is: ",deck)
```
